### Imoprt library Pandas and use method read_html

```python
import pandas as pd
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header = 0)
type(tables)
```

```python
len(tables)
```

```python
df = tables[0]
df.head()
```

```python
df.info()
```

### Delete rows that contain a value = 'Not assigned'

```python
indexNames = df[df['Borough'] == 'Not assigned'].index
df.drop(indexNames, inplace = True)
df.head(10)
```

### Reset Index in Data Frame

```python
df.reset_index(drop=True, inplace=True)
df.head(10)
```

### Insert a coma instead a slash

```python
df['Neighborhood'] = df['Neighborhood'].str.replace(' /', ',', regex = True)
df.head(10)
```

### Check the Column 'Neighborhood' for value 'Not assigned'

```python
n = 0
for v in df['Neighborhood']:
    if v == 'Not assigned':
        n = n + 1
print(n)
        
```

### Read CSV file with geodata

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.head(20)
```

#### The data in the received data frame is sorted alphabetically by the Postal Code column.

### Sort Data Frame "df" by column "Postal code"

```python
df.sort_values(['Postal code', 'Borough', 'Neighborhood'], ascending = [True, False, False], inplace = True)
df.head()
```

### Reset Index again

```python
df.reset_index(drop=True, inplace=True)
df.head(10)
```

### Insert geodata into the data frame

```python
df['Latitude'] = df_geo['Latitude']
df['Longitude'] = df_geo['Longitude']
df.head(10)
```
